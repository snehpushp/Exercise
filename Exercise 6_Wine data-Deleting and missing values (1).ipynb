### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

### Step 3. Assign it to a variable called wine

```python
wine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data",header=None)
wine
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine.drop([0,3,6,8,10,12,13],axis=1)
```

### Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):  
1) Alcohol <br>
2) Malic acid<br>
3) Ash<br>
4) Alcalinity of ash<br>
5) Magnesium<br>
6) Total phenols<br>
7) Flavanoids<br>
8) Nonflavanoid phenols<br>
9) Proanthocyanins<br>
10)Color intensity<br>
11)Hue<br>
12)OD280/OD315 of diluted wines<br>
13)Proline<br><br>

NOTE: 0th attribute is class identifier - Type of cultivator (1-3)<br>



```python
rename = {
    0 : "Type of cultivator",
    1 : "Alcohol",
    2 : "Malic acid",
    3 : "Ash",
    4 : "Alcalinity of ash",
    5 : "Magnesium",
    6 : "Total phenols",
    7 : "Flavanoids",
    8 : "Nonflavanoid phenols",
    9 : "Proanthocyanins",
    10 : "Color intensity",
    11 : "Hue",
    12 : "OD280/OD315 of diluted wines",
    13 : "Proline"
}
wine.rename(columns = rename,inplace = True,)
wine
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine['Alcohol'][0:3] = np.nan
wine
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine["Magnesium"][[3,4]] = np.nan
wine
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine.Alcohol[wine.Alcohol.isnull()] = 10
wine.Magnesium[wine.Magnesium.isnull()] = 100
wine
```

### Step 9. Count the number of missing values

```python
count = 0
for col in wine:
    for cell in wine[col]:
        if cell == np.nan:
            count += 1
count
```

### Step 10.  Create an array of 10 random numbers up until 10

```python
random_numbers = np.random.randint(0,10,10)
random_numbers
```

### Step 11.  Use random numbers you generated as an index and assign NaN value to each of cell.

```python
for i in random_numbers:
    wine.loc[wine.index == i] = np.nan
wine.head(10)
```

### Step 12.  How many missing values do we have?

```python
count = 0
for col in wine:
    for cell in wine[col]:
        if np.isnan(cell):
            count += 1
count
```

### Step 13. Delete the rows that contain missing values

```python
wine.dropna(inplace=True)
wine
```

### Step 14. Print only the non-null values in alcohol

```python
wine.Alcohol[np.isnan(wine.Alcohol)]
```

### Step 15.  Reset the index, so it starts with 0 again

```python
wine = wine.reset_index()
wine
```
